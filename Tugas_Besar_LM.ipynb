```python
# Sistem Pendukung Keputusan Penerimaan Karyawan

def keputusan(a, b, c):
    return (a and b) or (b and c)

print("SISTEM PENERIMAAN KARYAWAN")
print("1 = Ya, 0 = Tidak")

A = int(input("Tes tertulis ≥ 75? "))
B = int(input("Wawancara baik? "))
C = int(input("Pengalaman kerja? "))

hasil = keputusan(A, B, C)

if hasil:
    print("KEPUTUSAN: DITERIMA")
else:
    print("KEPUTUSAN: DITOLAK")
```
